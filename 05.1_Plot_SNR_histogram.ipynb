```python
import pandas as pd
import glob, os 
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = '/Volumes/home/Research/Detect_fish/Taoyuan_csv/23082011'
```

```python
allSNR = pd.DataFrame()
for file in glob.glob(f'{path}/*csv'):
    df = pd.read_csv(file)
    df_ = pd.concat([df['SNR'], df['SPLpp'], df['low_freq']], axis=1)
    allSNR = pd.concat([allSNR, df_], axis=0)
    
    
```

```python
plt.figure(figsize=(7,5))
bins = np.arange(-0.5,8.5,0.5)
n, _, patches = plt.hist(allSNR['SNR'], bins=bins, width = 0.5, 
                           edgecolor='white', color='black', align='mid')
for x, y in zip(bins[:-1], n):
    if int(y) >0:
        plt.text(x=x+0.05, y=y, s=int(y), fontsize=7)
        
plt.xlabel('SNR', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(bins)
plt.title(f"Taoyuan {path.rsplit('/')[-1]}", fontsize=12)
```
